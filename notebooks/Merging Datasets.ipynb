```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
good_reads = pd.read_csv('./goodreads_books.csv')
good_reads.head()
```

```python
good_reads.info()
```

### Dropping unwanted columns

```python
good_reads.drop(columns = ['id', 'link', 'cover_link', 'author_link', 'date_published', 'original_title', 'asin', 'settings',
                'characters', 'amazon_redirect_link', 'worldcat_redirect_link', 'recommended_books', 'books_in_series',
                'isbn', 'isbn13'], axis = 1,
               inplace=True)
```

```python
#good_reads['is_series'] = good_reads.series.notnull().astype('int')
#good_reads['has_won_award'] = good_reads.awards.notnull().astype('int')
#initialize target column
good_reads['nyt_best_seller'] = np.zeros(len(good_reads)).astype(int)
#good_reads.drop(columns=['series', 'awards'], axis = 1, inplace = True)
good_reads.head()
```

### Dropping Missing Values and Duplicates

```python
good_reads = good_reads.dropna(axis=0, subset=['publisher', 'genre_and_votes', 'description']).reset_index()
good_reads.isnull().sum().sort_values(ascending=False)
```

```python
good_reads['number_of_pages'].fillna(good_reads['number_of_pages'].mean(), inplace=True)
```

```python
good_reads.drop_duplicates(subset='title', keep='first', inplace=True)
good_reads.info()
```

```python
nyt = pd.read_pickle('full_nyt.pkl')
nyt.drop_duplicates('title', keep='first', inplace=True)
nyt.info()
```

```python
nyt2 = pd.read_json('./nyt2.json', lines=True)
nyt2.drop(columns=['_id', 'bestsellers_date', 'published_date', 'amazon_product_url',
                   'price', 'rank', 'rank_last_week', 'weeks_on_list'], inplace=True)
nyt2.drop_duplicates(subset='title', keep='first', inplace=True)
nyt2.info()
```

```python
good_reads['title'] = [title.lower() for title in good_reads.title.tolist()]
nyt['title'] = [title.lower() for title in nyt.title.tolist()]
nyt2['title'] = [title.lower() for title in nyt2.title.tolist()]
```

```python
matched = 0
nyt_notin_gr = 0 
for title in nyt2.title.tolist():
    if title in good_reads.title.tolist():
        good_reads.loc[good_reads.title == title, 'nyt_best_seller'] = 1
        matched += 1
    else:
        nyt_notin_gr += 1
```
