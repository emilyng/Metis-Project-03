```python
import pandas as pd
import json
```

```python
with open('nyt_books4.json') as json_file:
    data = json.load(json_file)
```

```python
dfs = []
for books in data:
    try:
        dfs.append(pd.json_normalize(books, ['results']))
    except:
        pass
```

```python
df = pd.concat(dfs)
df.info()
```

```python
df.head()
```

```python
#drop duplicates
new_df = df.drop_duplicates(subset='title', keep='first').reset_index().drop('index', axis=1)
```

```python
new_df.info()
```

```python
new_df.drop(columns=['contributor', 'contributor_note', 'price', 'age_group', 
                     'ranks_history', 'reviews'])
```

```python
pd.to_pickle(new_df, 'full_nyt.pkl')
```
